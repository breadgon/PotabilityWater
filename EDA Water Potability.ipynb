# Exploratory Data Analyst Water Potability

```python
#Import Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
water = pd.read_csv('water_potability.csv')
water
```

```python
water.info()
```

```python
water.isnull().sum()
```

```python
water.drop(columns=['Sulfate'], inplace=True)
```

```python
water
```

```python
# Fill NaN with forward and backward data

water.fillna(method='pad', inplace=True)
water.fillna(method='bfill', inplace=True)
```

```python
water
```

```python
water.isnull().sum()
```

```python
# Use Scatter Plot

plt.figure(figsize=(15,5))
plt.scatter(water['Trihalomethanes'], water['Chloramines'])

plt.xlabel('Trihalomethanes')
plt.ylabel('Chloramines')
plt.title('Trihalomethanes Vs Chloramines')
plt.show()
```

```python
plt.style.use('classic')

plt.figure(figsize=(15,7))
plt.scatter(water['ph'], water['Organic_carbon'], c=water['Potability'], s=75)

plt.xlabel('Ph')
plt.ylabel('Organic Carbon')
plt.title('Ph Vs Organic Carbon')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(15,8))


plt.subplot(3,1,1)
plt.scatter(water['ph'], water['Chloramines'])
plt.ylabel('Chloramines')


plt.subplot(3,1,2)
plt.scatter(water['ph'], water['Organic_carbon'])
plt.xlabel('ph')
plt.ylabel('Organic_carbon')


plt.subplot(3,1,3)
plt.scatter(water['ph'], water['Turbidity'])
plt.xlabel('ph')
plt.ylabel('Turbidity')

plt.show()
```

```python
for i in range(len(water.columns)):
    plt.boxplot(water.iloc[:, i])
    x_name = water.columns[i]
    plt.xlabel(f'{x_name}')
    plt.show()
```

```python
for i in range(len(water.columns)):
    plt.hist(water.iloc[:, i])
    x_name = water.columns[i]
    plt.title(f'{x_name}')
    plt.show()
```

```python
var_potability = water['Potability'].value_counts()
var_potability
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(9, 8))

labels = ['Non_Potable', 'Potable']
ax.pie(var_potability, labels = labels, shadow = True, autopct='%.2f%%', startangle=70)
ax.set_title('Potability')
ax.legend()

plt.show()
```

```python
water_potable = water[water['Potability'] == 1]
water_potable
```

```python
water_potable = water_potable.mean()
water_potable
```

```python
water_potable['Solids'] = water_potable['Solids']/100
water_potable
```

```python
water_potable = pd.DataFrame(water_potable, columns = ['Potability'])
water_potable
```

```python
water_potable = water_potable.drop(['Potability'])
water_potable
```

```python
plt.style.use('fivethirtyeight')
water_potable.plot(kind='bar');
```

```python
water_non_potable = water[water['Potability'] == 0]
water_non_potable
```

```python
water_non_potable = water_non_potable.mean()
water_non_potable
```

```python
water_non_potable['Solids'] = water_non_potable['Solids']/100
water_non_potable
```

```python
water_non_potable = pd.DataFrame(water_non_potable, columns=['Non Potability'])
```

```python
water_non_potable
```

```python
water_non_potable = water_non_potable.drop(['Potability'])
```

```python
water_non_potable
```

```python
plt.style.use('fivethirtyeight')
water_non_potable.plot(kind='bar');
```

```python
concat_water = pd.concat([water_potable, water_non_potable], axis=1)
concat_water
```

```python
plt.style.use('default')

width = 0.4
x = np.arange(len(concat_water.index))

fig1, ax1 = plt.subplots(figsize=(12,5))
reacts1 = ax1.bar(x - width/2, concat_water['Potability'], width, label = 'Potable')
reacts2 = ax1.bar(x + width/2, concat_water['Non Potability'], width, label = 'Non Potable')
ax1.set_title('Potability of Water', pad=25, fontweight='bold')
ax1.set_ylabel('Potability')
ax1.set_xticks(x)
ax1.set_xticklabels(concat_water.index, rotation=30)
ax1.legend()

ax1.bar_label(reacts1, padding=2, rotation=30)
ax1.bar_label(reacts2, padding=2, rotation=30)

fig1.tight_layout()

plt.show()
```
